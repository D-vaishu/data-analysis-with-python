```python
import selenium as se
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
options=webdriver.ChromeOptions()
```

```python
driver=webdriver.Chrome(options=options)
```

```python
driver.get("https://www.myntra.com/")
```

```python
search = driver.find_element(By.XPATH,".//input[@class ='desktop-searchBar']")
```

```python
search.send_keys("smart watches")
```

```python
search.send_keys(Keys.ENTER)
```

```python
colors=driver.find_elements(By.XPATH,".//span[@class='vertical-filters-header']")
color=[i.text for i in colors]
print(color)
```

```python
brand = driver.find_elements(By.XPATH,".//h3[@class='product-brand']")
brands=[i.text for i in brand]
print(brands)
```

```python
product=driver.find_elements(By.XPATH,".//h4[@class='product-product']")
products=[i.text for i in product]
print(products)
```

```python
df=pd.DataFrame(columns=["Brands","product"])
df
```

```python
df['Brands']=brands
df['product']=products
df
```

```python
df.to_csv("mytra.csv")
```
