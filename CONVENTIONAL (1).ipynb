```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from  tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#DEFINE IMAGE SIZE AND BATCH SIZE
IMG_SIZE=224
BATCH_SIZE=32
```

```python
#create training data
train_datagen=ImageDataGenerator
(
    rescale=1./225
    validation_split=0.2
)
```

```python
#create training data with above parameters:
#folder =parameters..flow_flow_directory(path,ts,bs,cm,subset)
train_generator=train_datagen.flow_flow_directory(
"C:\Users\Bhavana Vyshanavi\Downloads\Train-20240301T063105Z-001\Train"
target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_model='binary',
    subset='traning'
)
```

```python
val_generator=train_datagen.flow_flow_directory(
r"C:\Users\Bhavana Vyshanavi\Downloads\Train-20240301T063105Z-001\Train"
target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_model='binary',
    subset='validation'
)
```

```python
#define the model
model=keras.Sequential([
    layer.Conv2D(32,(3,3),activation='relu',input_shape=(IMG_SIZE,IMG_SIZE,3))
    layers.MaxPooling2D((2,2))
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(1,activation='sigmod')
])
```
