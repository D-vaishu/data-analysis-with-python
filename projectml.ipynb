```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
feature_cols=['Juiciness','Weight','Crunchiness']
features=df[feature_cols]
Target=df.Quality
```

```python
RF=RandomForestClassifier(n_estimators=25)
RF.fit(features,Target)
```

```python
import os
output_dir="tree_visualizations"
os.makedirs(output_dir,exist_ok=True)
```

```python
RF.estimators_
```

```python
for i,tree in enumerate(RF.estimators_):
  tree_dot_file=os.path.join(output_dir,f"true_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"true_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file= tree_dot_file, feature_names=["Juiciness", "Weight", "Crunchiness"],
class_names=[str(cls) for cls in RF.classes_], filled=True, rounded=True)
```

```python
command=f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree{i} visualization saved to {tree_png_file}")
```
